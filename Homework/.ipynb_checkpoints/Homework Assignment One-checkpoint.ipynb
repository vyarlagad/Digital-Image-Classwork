```python
%matplotlib inline
import cv2
import matplotlib.pyplot as plt
import os
```

```python
# Obtain the Two Images (Uses Absolute Path)
img = cv2.imread("/Users/Vasukesh/Desktop/Digital-Image-Classwork/Images/rare_pupper.jpg",1) # 1 Indicates Full Color
img2 = cv2.imread("/Users/Vasukesh/Desktop/Digital-Image-Classwork/Images/rare_pupper.png",1)
# Setup Subplots
plt.figure(num=None, figsize=(18, 16), dpi=80, facecolor='w', edgecolor='k')
plt.subplot(121)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

plt.subplot(122)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Compare File Sizes
print('File Size in Bytes: ', os.stat('/Users/Vasukesh/Desktop/Digital-Image-Classwork/Images/rare_pupper.jpg').st_size)
print('File Size in Bytes: ', os.stat('/Users/Vasukesh/Desktop/Digital-Image-Classwork/Images/rare_pupper.png').st_size)
```
