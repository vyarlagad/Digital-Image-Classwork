```python
%matplotlib inline
import cv2
import matplotlib.pyplot as plt
import os
```

```python
# Set Image Path
img_path = "/Users/Vasukesh/Desktop/Digital-Image-Classwork/Images/rare_pupper.jpg"
img2_path = "/Users/Vasukesh/Desktop/Digital-Image-Classwork/Images/rare_pupper.png"
# Obtain the Two Images (Uses Absolute Path)
img = cv2.imread(img_path,1) # 1 Indicates Full Color
img2 = cv2.imread(img2_path,1)
# Setup Subplots
plt.figure(num=None, figsize=(18, 16), dpi=80, facecolor='w', edgecolor='k')
plt.subplot(121)
plt.title('Trash Panda JPG')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

plt.subplot(122)
plt.title('Trash Panda PNG')
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Compare File Sizes
print('File Size in Bytes: ', os.stat(img_path).st_size)
print('File Size in Bytes: ', os.stat(img2_path).st_size)
```

### Description:
** The two image file sizes really show the difference between the PNG and Jpeg. <br>
   Jpeg uses lossy compression so as this figure starts to increase in size then so does the quality of the image. <br>
   PNG on the otherhand, uses lossless so the quality wont fall off too far however to compensate there is a much larger file size.
